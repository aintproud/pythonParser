```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# r = requests.get('https://www.imdb.com/chart/top/')
# html = r.text
# print(html)
# with open('top250.html', 'w', encoding='utf-8') as f:
#     f.write(html)
with open('top250.html', 'r', encoding='utf-8') as f:
    html = f.read()
soup = BeautifulSoup(html, 'html.parser')
```

```python
films_trs = soup.find_all('tr')
films_trs.pop(0)

places = []
names = []
years =[]
raitings =[]
```

```python
for film_tr in films_trs:
    some_info = film_tr.find(class_="titleColumn").text.strip('\n').split('\n')
    place = int(float(some_info[0]))
    name = some_info[1].strip()
    year = int(some_info[2][1:-1])
    raiting = float(film_tr.find(class_="ratingColumn imdbRating").text.strip())
    print('\n', place, name, year, raiting)
    places.append(place)
    names.append(name)
    years.append(year)
    raitings.append(raiting)
```

```python

df = pd.DataFrame({'place': places, 'name':names, 'year': years, 'raiting': raitings})
df = df.set_index('place')
print(df.head(5))
```
